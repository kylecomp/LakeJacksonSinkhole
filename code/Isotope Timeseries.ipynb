```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import Data
precip_df = pd.read_csv('../data/LJPrecip2021.csv', parse_dates=['Timestamp (UTC)'],
                        index_col='Timestamp (UTC)', squeeze=True)

zoom=precip_df.loc['20210601':'20211031']

```

```python
iso_df = pd.read_csv('../data/Isotope.csv', parse_dates=['Sampling time'],
                        index_col='Sampling time', squeeze=True)
```

```python
LJ=iso_df.loc[iso_df['Location'] == 'Porter Sink']
FW=iso_df.loc[iso_df['Location'] == 'FSU Well']
WS=iso_df.loc[iso_df['Location'] == 'Wakulla Springs']
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(4,1,1)
plt.plot(LJ.index,LJ['d18O vs vsmow'], 'o-')
plt.title('Lake Jackson d18O')
#plt.xlim(['2021-06-01', '2021-10-31' ])
plt.xticks(rotation=30)
#Sinkhole Exposed Lines
plt.vlines(('2021-06-06', '2021-07-05'), ymin=-1,ymax=6, colors='r')
#Sinkhole Refilled Lines
plt.vlines(('2021-06-23', '2021-08-16'), ymin=-1,ymax=6, colors='k')

plt.subplot(4,1,2)
plt.bar(zoom.index,zoom['Value (in)'])
plt.title('Lake Jackson daily Precip')
#plt.xlim(['2021-06-01', '2021-10-31' ])
plt.xticks(rotation=30)
plt.vlines(('2021-06-06', '2021-07-05'), ymin=0,ymax=0.8, colors='r')
#Sinkhole Refilled Lines
plt.vlines(('2021-06-23', '2021-08-16'), ymin=0,ymax=0.8, colors='k')
#plt.vlines(('2021-07-18'), ymin=0,ymax=0.8, colors='g')
plt.subplot(4,1,3)
plt.plot(FW.index, FW['d18O vs vsmow'], 'o-')
plt.title('FSU Well d18O')
#plt.xlim(['2021-06-01', '2021-10-31' ])
plt.xticks(rotation=30)
plt.vlines(('2021-06-06', '2021-07-05'), ymin=-3.8,ymax=-3, colors='r')
#Sinkhole Refilled Lines
plt.vlines(('2021-06-23', '2021-08-16'), ymin=-3.8,ymax=-3, colors='k')

plt.subplot(4,1,4)
plt.plot(WS.index, WS['d18O vs vsmow'], 'o-')
plt.title('Wakulla Springs d18O')
#plt.xlim(['2021-06-01', '2021-10-31' ])
plt.xticks(rotation=30)
plt.vlines(('2021-06-06', '2021-07-05'), ymin=-3.8,ymax=-3, colors='r')
#Sinkhole Refilled Lines
plt.vlines(('2021-06-23', '2021-08-16'), ymin=-3.8,ymax=-3, colors='k')


plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=1, 
                    wspace=0.4, 
                    hspace=0.4)
plt.show()
```

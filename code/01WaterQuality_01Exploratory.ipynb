```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Import Recipitation Data
precip_df = pd.read_csv('../data/LJPrecip2021.csv', parse_dates=['Timestamp (UTC)'],
                        index_col='Timestamp (UTC)', squeeze=True)

zoom=precip_df.loc['20210201':'20211031']

wq_df = pd.read_csv('../data/MasterWaterQuality.csv', parse_dates=['Sampling time'],
                        index_col='Sampling time', squeeze=True)
well_df=pd.read_csv('../data/processed/Well2021.csv', parse_dates=['Datetime'],
                        index_col='Datetime', squeeze=True)
```

```python
LJ=wq_df.loc[wq_df['Location'] == 'Porter Sink']
FW=wq_df.loc[wq_df['Location'] == 'FSU Well']
WS=wq_df.loc[wq_df['Location'] == 'Wakulla Springs']
SW=wq_df.loc[wq_df['Location'] == 'Sally Ward Spring']
```

```python
fig, ax = plt.subplots(4,1, figsize=(15,15))
fig.tight_layout(pad=3.0)
ax[0].plot(LJ.index, LJ.Calcium,'-o',color='k')
ax[0].vlines(('2021-06-06', '2021-07-05'), ymin=-1,ymax=6, colors='r')
ax[0].vlines(('2021-06-23', '2021-08-16'), ymin=-1,ymax=6, colors='b')
ax2=ax[0].twinx()
ax2.plot(zoom.index, zoom["Value (in)"])
ax[0].title.set_text('Lake Jackson Calcium')

ax[1].plot(FW.index, FW.Calcium,'-o',color='k')
ax[1].vlines(('2021-06-06', '2021-07-05'), ymin=-1,ymax=80, colors='r')
ax[1].vlines(('2021-06-23', '2021-08-16'), ymin=-1,ymax=80, colors='b')
ax2=ax[1].twinx()
ax2.plot(zoom.index, zoom["Value (in)"])
ax[1].title.set_text('FSU Well Calcium')

ax[2].plot(WS.index, WS.Calcium,'-o',color='k')
ax[2].vlines(('2021-06-06', '2021-07-05'), ymin=40,ymax=50, colors='r')
ax[2].vlines(('2021-06-23', '2021-08-16'), ymin=40,ymax=50, colors='b')
ax2=ax[2].twinx()
ax2.plot(zoom.index, zoom["Value (in)"])
ax[2].title.set_text('Wakulla Springs Calcium')

ax[3].plot(SW.index, SW.Calcium,'-o',color='k')
ax[3].vlines(('2021-06-06', '2021-07-05'), ymin=40,ymax=50, colors='r')
ax[3].vlines(('2021-06-23', '2021-08-16'), ymin=40,ymax=50, colors='b')
ax2=ax[3].twinx()
ax2.plot(zoom.index, zoom["Value (in)"])
ax[3].title.set_text('Sally Ward Springs Calcium')
```

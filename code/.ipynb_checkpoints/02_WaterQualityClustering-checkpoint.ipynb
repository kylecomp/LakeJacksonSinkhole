```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, cluster
from sklearn.preprocessing import StandardScaler
from scipy.spatial import distance
import scipy
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
wq_df = pd.read_csv('../data/MasterWaterQuality.csv', parse_dates=['Sampling time'],
                        index_col='Sampling time', squeeze=True)
```

```python
wq=wq_df.loc[:, ~wq_df.columns.isin(['color', 'pH', 'pOH', 'Total Alkalinity (mg/L as CaCO3)','Bicarbonate', 'EC(us/cm)'])]
NoNa=wq.dropna()
chemicals=NoNa.loc[:, ~NoNa.columns.isin(['ID', 'Location'])]
loc=NoNa['ID']
X= StandardScaler().fit_transform(chemicals)
loc[0:10]
```

```python
X.shape
```

```python
#B Create Linkage Matrix and Plot Dendrogram
Z = linkage(chemicals, 'single')
#fig = plt.figure(figsize=(20,10))
#dn = dendrogram(Z, labels=labels)
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1)
dendrogram(Z, ax=ax,labels=loc, orientation='top')
ax.tick_params(axis='x', which='major', labelsize=8)
ax.tick_params(axis='y', which='major', labelsize=8)
fig.savefig('../plots/single2.png')
plt.show()

```

```python
#B Create Linkage Matrix and Plot Dendrogram
#B Create Linkage Matrix and Plot Dendrogram
Z = linkage(chemicals, 'complete')
#fig = plt.figure(figsize=(20,10))
#dn = dendrogram(Z, labels=labels)
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1)
dendrogram(Z, ax=ax,labels=loc, orientation='top')
ax.tick_params(axis='x', which='major', labelsize=8)
ax.tick_params(axis='y', which='major', labelsize=8)
fig.savefig('../plots/complete2.png')
plt.show()
```

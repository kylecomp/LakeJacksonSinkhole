```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mplcursors
from matplotlib.lines import Line2D
from matplotlib.animation import FuncAnimation, PillowWriter
from IPython.display import HTML
from sklearn.linear_model import LinearRegression
```

```python
#Import Data
precip_df = pd.read_csv('../data/LJPrecip2021.csv', parse_dates=['Timestamp (UTC)'],
                        index_col='Timestamp (UTC)', squeeze=True)

zoom=precip_df.loc['20210601':'20211031']

wq_df = pd.read_csv('../data/MasterWaterQuality.csv', parse_dates=['Sampling time'])
wq_df
```

```python
iso_df = pd.read_csv('../data/Isotope.csv', parse_dates=['Sampling time'],
                        index_col='Sampling time', squeeze=True)
```

```python
LJ=iso_df.loc[iso_df['Location'] == 'Porter Sink']
FW=iso_df.loc[iso_df['Location'] == 'FSU Well']
WS=iso_df.loc[iso_df['Location'] == 'Wakulla Springs']
precip=iso_df.loc[(iso_df['Location'] == 'T.S. Fred') | (iso_df['Location'] == 'Rain')]
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(4,1,1)
plt.plot(LJ.index,LJ['d18O vs vsmow'], 'o-')
plt.title('Lake Jackson d18O')
#plt.xlim(['2021-06-01', '2021-10-31' ])
plt.xticks(rotation=30)
#Sinkhole Exposed Lines
plt.vlines(('2021-06-06', '2021-07-05'), ymin=-1,ymax=6, colors='r')
#Sinkhole Refilled Lines
plt.vlines(('2021-06-23', '2021-08-16'), ymin=-1,ymax=6, colors='k')

plt.subplot(4,1,2)
plt.bar(zoom.index,zoom['Value (in)'])
plt.title('Lake Jackson daily Precip')
#plt.xlim(['2021-06-01', '2021-10-31' ])
plt.xticks(rotation=30)
plt.vlines(('2021-06-06', '2021-07-05'), ymin=0,ymax=0.8, colors='r')
#Sinkhole Refilled Lines
plt.vlines(('2021-06-23', '2021-08-16'), ymin=0,ymax=0.8, colors='k')
#plt.vlines(('2021-07-18'), ymin=0,ymax=0.8, colors='g')
plt.subplot(4,1,3)
plt.plot(FW.index, FW['d18O vs vsmow'], 'o-')
plt.title('FSU Well d18O')
#plt.xlim(['2021-06-01', '2021-10-31' ])
plt.xticks(rotation=30)
plt.vlines(('2021-06-06', '2021-07-05'), ymin=-3.8,ymax=-3, colors='r')
#Sinkhole Refilled Lines
plt.vlines(('2021-06-23', '2021-08-16'), ymin=-3.8,ymax=-3, colors='k')

plt.subplot(4,1,4)
plt.plot(WS.index, WS['d18O vs vsmow'], 'o-')
plt.title('Wakulla Springs d18O')
#plt.xlim(['2021-06-01', '2021-10-31' ])
plt.xticks(rotation=30)
plt.vlines(('2021-06-06', '2021-07-05'), ymin=-3.8,ymax=-3, colors='r')
#Sinkhole Refilled Lines
plt.vlines(('2021-06-23', '2021-08-16'), ymin=-3.8,ymax=-3, colors='k')


plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=1, 
                    wspace=0.4, 
                    hspace=0.4)
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5)
ax.plot(WS.index, WS['d18O vs vsmow'], 'o-')
ax.set_title('Wak Spring d18O')
#ax.set_xlim(['2021-06-01', '2021-10-31' ])
#ax.xticks(rotation=30)
ax.vlines(('2021-06-06', '2021-07-05'), ymin=-3.8,ymax=-3, colors='r')
#Sinkhole Refilled Lines
ax.vlines(('2021-06-23', '2021-08-16'), ymin=-3.8,ymax=-3, colors='k')
plt.show()
```

```python
y=precip['dD vs vsmow'].to_numpy()
y=y.reshape(-1,1)
x=precip['d18O vs vsmow'].to_numpy()
x=x.reshape(-1,1)
model=LinearRegression().fit(x,y)

lmwl_x=np.linspace(-7.5,7.5,50)
lmwl_y=model.coef_ * lmwl_x + model.intercept_
lmwl_y=lmwl_y.reshape(-1,1)
print("LMWL Equation: d18O= ", model.coef_, "* dD +", model.intercept_ , "with R2 = ", model.score(x, y))
```

```python
fig, ax = plt.subplots()
y=wq_df['dD vs vsmow']
x=wq_df['d18O vs vsmow']

fig.set_size_inches(18.5, 10.5)

ax.scatter(x,y,c=wq_df['color'], )
ax.plot(lmwl_x,lmwl_y)
ax.set_ylabel('dD vs VSMOW')
ax.set_xlabel('d18O vs VSMOW')
legend_elements = [Line2D([0], [0], marker='o', color='k', label='Porter Sink', markersize=10),
                   Line2D([0], [0], marker='o', color='r', label='FSU Well',  markersize=10),
                   Line2D([0], [0], marker='o', color='b', label='Blue Sink', markersize=10),
                   Line2D([0], [0], marker='o', color='g', label='Munson Slough', markersize=10),
                   Line2D([0], [0], marker='o', color='c', label='Sally Ward',markersize=10),
                   Line2D([0], [0], marker='o', color='m', label='Wakulla', markersize=10),
                   Line2D([0], [0], marker='o', color='y', label='Rain', markersize=10)]
            
                   

ax.legend(handles=legend_elements)

mplcursors.cursor(hover=True)
plt.show()
```

```python
Well=FW.to_numpy()

fig, ax = plt.subplots(1,1, figsize=(10,10))

ax.set_ylabel('dD vs vsmow')
ax.set_xlabel('d18O vs VSMOW')
ax.set_ylim(-25, 0)
ax.set_xlim(-5, -2)
ax.plot(lmwl_x,lmwl_y)

label = ax.text(10, 1.1, FW.index[0], ha='center', va='center', fontsize=20, color="Red")

def animation(i):
    ax.scatter(Well[i,3], Well[i,2], s=50, c='k')
    label.set_text(FW.index[i])


ani = FuncAnimation(fig, animation, interval=1000, frames= 27)
HTML(ani.to_jshtml())
```
